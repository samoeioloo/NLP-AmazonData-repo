```python
#Automate the 3 above assignments
#1: Apply bag of words
```

```python
def textFit(x,y,nlpModel,mlModel,coef_show=1):
    xC = nlpModel.fit_transform(x)
    print('features:{}'.format(xC.shape[1]))
    xTrain, xTest, yTrain, yTest = train_test_split(xC, y) 
    #apply machine learning model
    ml=mlModel.fit(xTrain, yTrain)
    acc=ml.score(xTest, yTest) #2: checking accuracy
    print(acc)

    if (coef_show==1):
        #printing top 20 pos and neg words
        w = c.get_feature_names() 
        coef = ml.coef_.tolist()[0]
        coefDF = pd.DataFrame({'Word': w, 'Coefficient':coef})
        coefDF =coefDF.sort_values(['Coefficient', 'Word'], ascending=False) 
        print('\n')
        print("Top 20 Positive words \n")
        print(coefDF.head(20))
        print('\n')
        print("Tope 20 Negative words \n")
        print(coefDF.tail(20))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
c=CountVectorizer(stop_words='english')
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
#call function
textFit(x,y,c,LogisticRegression())
```

```python
#automate the predictions
from sklearn.metrics import confusion_matrix, accuracy_score
def predict(x,y,nlpModel, mlModel):
    xC=nlpModel.fit_transform(x)
    #perform train test split
    xTrain, xTest, yTrain, yTest = train_test_split(xC, y)
    ml=mlModel.fit(xTrain,yTrain)
    predictions=ml.predict(xTest)
    cm=confusion_matrix(predictions,yTest)
    print(cm)
    acc=accuracy_score(predictions,yTest)
    print(acc)
    
```

```python
c=CountVectorizer()
lr=LogisticRegression()
```

```python
predict(x,y,c,lr)
```
